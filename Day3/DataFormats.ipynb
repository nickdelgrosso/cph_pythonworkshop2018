## SQLite 

```python
import sqlalchemy
engine = sqlalchemy.create_engine('sqlite:///Mydata.db')
```

```python
conn = engine.connect()
```

```python
df.to_sql('Proteins', conn)
```

```python
pd.read_sql_table('Proteins', conn).head()
```

```python
query = 'SELECT respcolor, stimcolor FROM Proteins WHERE respcolor = "Y"'
dd = pd.read_sql_query(query, conn)
dd.head()
```

```python
dd.to_sql('SomeProteins', conn)
```
